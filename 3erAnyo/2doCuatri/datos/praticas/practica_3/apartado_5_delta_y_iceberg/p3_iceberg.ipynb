# Experiment: Practica 3 Iceberg

Este notebook crea `iceberg.gold.dim_category`, repite las operaciones del apartado Delta y compara ambos formatos.


## Preparacion

Se crea el namespace Iceberg y se construye la tabla a partir de la dimension Delta ya generada por el pipeline.


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("p3_iceberg").getOrCreate()

DELTA_PATH = "s3a://lakehouse/gold/dim_category"
ICEBERG_TABLE = "iceberg.gold.dim_category"
TEST_KEY = "p3_test_dim_category_20260319"

spark.sql("CREATE NAMESPACE IF NOT EXISTS iceberg.gold")

delta_dim_category = spark.read.format("delta").load(DELTA_PATH)
delta_dim_category.writeTo(ICEBERG_TABLE).using("iceberg").createOrReplace()

spark.sql(f"SELECT * FROM {ICEBERG_TABLE} LIMIT 5").show(truncate=False)
```

## 1. Insercion del registro ficticio


```python
spark.sql(f"""
INSERT INTO {ICEBERG_TABLE} VALUES (
    '{TEST_KEY}',
    'P3_TEST_CATEGORY',
    'P3_TEST_SUBCATEGORY',
    'P3 inserted row',
    'P3_TEST_REPORT'
)
""")

spark.sql(f"SELECT * FROM {ICEBERG_TABLE} WHERE category_key = '{TEST_KEY}'").show(truncate=False)
```

## 2. Modificacion del registro insertado


```python
spark.sql(f"""
UPDATE {ICEBERG_TABLE}
SET incident_description = 'P3 updated row',
    report_type_code = 'P3_TEST_REPORT_UPDATED'
WHERE category_key = '{TEST_KEY}'
""")

spark.sql(f"SELECT * FROM {ICEBERG_TABLE} WHERE category_key = '{TEST_KEY}'").show(truncate=False)
```

## 3. Eliminacion del registro insertado


```python
spark.sql(f"DELETE FROM {ICEBERG_TABLE} WHERE category_key = '{TEST_KEY}'")

spark.sql(f"SELECT * FROM {ICEBERG_TABLE} WHERE category_key = '{TEST_KEY}'").show(truncate=False)
```

## Historial y time travel


```python
history = spark.sql(f"SELECT * FROM {ICEBERG_TABLE}.history ORDER BY made_current_at DESC")
snapshots = spark.sql(f"SELECT * FROM {ICEBERG_TABLE}.snapshots ORDER BY committed_at DESC")

history.show(truncate=False)
snapshots.show(truncate=False)

snapshot_rows = snapshots.select("snapshot_id").collect()
for row in snapshot_rows[:4]:
    snapshot_id = row["snapshot_id"]
    print(f"===== SNAPSHOT {snapshot_id} =====")
    (
        spark.read.format("iceberg")
        .option("snapshot-id", snapshot_id)
        .load(ICEBERG_TABLE)
        .where(f"category_key = '{TEST_KEY}'")
        .show(truncate=False)
    )
```

## Comparacion Delta vs Iceberg

### 1. Historial de cambios
- En Delta Lake el historial se consulta con `DESCRIBE HISTORY gold.dim_category`.
- En Iceberg el historial se consulta mediante tablas de metadatos como `iceberg.gold.dim_category.history` y `iceberg.gold.dim_category.snapshots`.

### 2. Acceso a versiones anteriores
- En Delta se accede con `versionAsOf` o `TIMESTAMP AS OF`.
- En Iceberg se accede con `snapshot-id` o `VERSION AS OF` sobre snapshots concretos.

### 3. Diferencias de sintaxis y configuracion
- Delta necesita registrar o leer la tabla en formato Delta y luego permite `INSERT`, `UPDATE` y `DELETE` directamente.
- Iceberg requiere usar un catalogo Iceberg (`iceberg.gold`) y la sintaxis habitual pasa por namespaces y snapshots.
- En este entorno Delta usa la ruta `s3a://lakehouse/gold/...` y Iceberg usa el catalogo configurado en `spark-defaults.conf`.

### 4. Diferencias en metadatos
- Delta centraliza el versionado en `_delta_log` con ficheros JSON y checkpoints.
- Iceberg expone mas informacion estructurada a traves de sus tablas de metadatos (`history`, `snapshots`, `files`, `manifests`).
- Iceberg facilita inspeccionar snapshots y manifests desde SQL, mientras que Delta concentra el detalle en el log transaccional.
