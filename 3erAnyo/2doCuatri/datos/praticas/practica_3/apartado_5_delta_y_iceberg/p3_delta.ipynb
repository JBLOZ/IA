# Experiment: Practica 3 Delta Lake

Este notebook realiza las operaciones pedidas sobre `gold.dim_category` en formato Delta.


## Preparacion

Se registra la tabla Delta a partir de la ruta generada por el pipeline Bronze -> Silver -> Gold.


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("p3_delta").getOrCreate()

DELTA_PATH = "s3a://lakehouse/gold/dim_category"
TEST_KEY = "p3_test_dim_category_20260319"

spark.read.format("delta").load(DELTA_PATH).show(5, truncate=False)
```

## 1. Insercion del registro ficticio


```python
spark.sql(f"""
INSERT INTO delta.`{DELTA_PATH}` VALUES (
    '{TEST_KEY}',
    'P3_TEST_CATEGORY',
    'P3_TEST_SUBCATEGORY',
    'P3 inserted row',
    'P3_TEST_REPORT'
)
""")

spark.read.format("delta").load(DELTA_PATH).where(f"category_key = '{TEST_KEY}'").show(truncate=False)
```

## 2. Modificacion del registro insertado


```python
spark.sql(f"""
UPDATE delta.`{DELTA_PATH}`
SET incident_description = 'P3 updated row',
    report_type_code = 'P3_TEST_REPORT_UPDATED'
WHERE category_key = '{TEST_KEY}'
""")

spark.read.format("delta").load(DELTA_PATH).where(f"category_key = '{TEST_KEY}'").show(truncate=False)
```

## 3. Eliminacion del registro insertado


```python
spark.sql(f"DELETE FROM delta.`{DELTA_PATH}` WHERE category_key = '{TEST_KEY}'")

spark.read.format("delta").load(DELTA_PATH).where(f"category_key = '{TEST_KEY}'").show(truncate=False)
```

## Historial y time travel


```python
history = spark.sql(f"DESCRIBE HISTORY delta.`{DELTA_PATH}`")
history.show(truncate=False)

current_version = history.agg({"version": "max"}).collect()[0][0]
versions_to_check = [version for version in [current_version, current_version - 1, current_version - 2, current_version - 3] if version is not None and version >= 0]

for version in versions_to_check:
    print(f"===== VERSION {version} =====")
    (
        spark.read.format("delta")
        .option("versionAsOf", version)
        .load(DELTA_PATH)
        .where(f"category_key = '{TEST_KEY}'")
        .show(truncate=False)
    )
```

## Observacion

Tras la operacion de borrado, el registro ya no aparece en la version actual, pero si puede consultarse en versiones anteriores gracias al versionado de Delta Lake.
