## Experimento aleatorio 

En este experimento se simula el lanzamiento de una moneda un número de veces determinado

```python
try:
    import texttable
except ImportError:
    !pip install texttable
    import texttable
    
import matplotlib.pyplot as plt
# Para crea tablas
from texttable import Texttable 

# Función para simular lanzamientos de moneda y obtener resultados
def lanzamiento_moneda(n):
    resultados = [random.choice(['cara', 'cruz']) for _ in range(n)]
    return resultados

# Simulación de lanzar la moneda 100 veces
resultados = lanzamiento_moneda(100)

# Contar la frecuencia de cada resultado
cara_count = resultados.count('cara')
cruz_count = resultados.count('cruz')



# Tabla Resumen
t1 = Texttable()
t1.add_rows([['Resultado','frecuencia', 'Porcentaje' ],
           ['cara',	cara_count, cara_count/(cara_count+cruz_count)*100],
           ['cruz',	cruz_count, cruz_count/(cara_count+cruz_count)*100]])
print(t1.draw())

# Generar histograma
plt.bar(['Cara', 'Cruz'], [cara_count, cruz_count], color = ['blue', 'salmon'])
plt.title('Histograma de Lanzamiento de Moneda')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()
```

## Definición de probabilidad.

Definición clásica: la probabilidad de un suceso $A$ es el cociente entre el número de casos favorables y el número de casos posibles.

Definición frecuentista (ley de grandes números): la probabilidad de un suceso es el valor límite de su frecuencia relativa al repetir indefinidamente la experimentación.

En teoría de probabilidad, podemos decir que dos eventos son independientes cuando la probabilidad de cada uno de ellos no se ve afecta porque el otro evento ocurra, es decir que no existe ninguna relación entre los eventos.

Se puede formular una "ley de grandes números" de la siguiente forma: si se repite un experimento aleatorio, bajo las mismas condiciones, un número ilimitado de veces; y si estas repeticiones son independientes la una de la otra, entonces la frecuencia de veces que un evento ocurra, convergerá a un número que es igual a la probabilidad de que ocurra en una sola repetición del experimento.

Para entender mejor la ley de grandes números, vamos a utilizar el ejemplo del lanzamiento de una moneda. En un primer ejemplo, la moneda va a tener la misma posibilidad de caer en cara o cruz. Mientras que en el segundo ejemplo, vamos a modificar la probabilidad de la moneda para que caiga cara solo en 1 de cada 6 veces.

```python
import matplotlib.pyplot as plt
# Ejemplo ley de grandes números
# moneda p=1/2 cara=1 seca=0
resultados = []
for lanzamientos in range(1,10000):
    lanzamientos = np.random.choice([0,1], lanzamientos)
    caras = lanzamientos.mean()
    resultados.append(caras)

# graficamente
fig, ax = plt.subplots()
width = 8
height = 6
fig.set_size_inches(width, height)
df = pd.DataFrame({ 'lanzamientos' : resultados})
ax.plot(df)
xlabel = ax. set_xlabel("Número de lanzamientos")
ylabel = ax.set_ylabel ("frecuencia de caras")
plt.show()
```

Otro gran teorema de la teoría de probabilidad es el Teorema del límite central. Este teorema establece que la suma o el promedio de casi cualquier conjunto de variables independientes generadas al azar se aproximan a la Distribución Normal. Este Teoremadel límite central explica por esta distribucion surge comúnmente y es, generalmente, una aproximación excelente para la media de casi cualquier colección de datos. Para ilustrar también este teorema, recurramos a un poco más de Python

```python
# Ejemplo teorema del límite central
muestra_binomial = []
muestra_exp = []
muestra_possion = []
muestra_geometric = []
mu = .9
lam = 1.0
size=1000

for i in range(1,20000):
    muestra = np.random.binomial(1, mu, size=size)
    muestra_binomial.append(muestra.mean())
    muestra = np.random.exponential(scale=2.0,size=size)
    muestra_exp.append(muestra.mean())
    muestra = np.random.geometric(p=.5, size=size)
    muestra_geometric.append(muestra.mean())
    muestra = np.random.poisson (lam=lam, size=size)
    muestra_possion.append(muestra.mean())

df = pd.DataFrame({ 'binomial' : muestra_binomial,
                     'poission' : muestra_possion,
                     'geometrica' : muestra_geometric,
                    'exponencial' : muestra_exp})

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
df.binomial.hist(ax=axes[0,0], alpha=0.9, bins=1000)
df.exponencial.hist(ax=axes[0,1],bins=1000)
df.poission.hist(ax=axes[1,0],bins=1000)
df.geometrica.hist(ax=axes[1,1],bins=1000)

axes[0,0].set_title('Binomial')
axes[0,1].set_title('Poisson')
axes[1,0].set_title('Geométrica')
axes[1,1].set_title('Exponencial')
plt.show()
```

### Ejemplo 2
Un concursante debe elegir una puerta entre tres (todas cerradas); el premio consiste en llevarse lo que se encuentra detrás de la elegida. Se sabe con certeza que tras una de ellas se oculta un ferrari, y tras las otras dos hay cabras. Una vez que el concursante haya elegido una puerta y comunicado su elección a los presentes, el presentador, que sabe lo que hay detrás de cada puerta, abrirá una de las otras dos, en la que habrá una cabra. A continuación, le da la opción al concursante de cambiar, si lo desea, de puerta (tiene dos opciones). ¿Debe el concursante mantener su elección original o escoger la otra puerta? ¿Hay alguna diferencia?

Intuitivamente, es bastante sencillo que nuestra elección original (cuando hay tres puertas para elegir) tiene una probabilidad de 1/3 de ganar el concurso. Las cosas sin embargo se complican, cuando se descarta una puerta. Muchos dirían que ahora tenemos una probabilidad de 1/2 de ganar, seleccionando cualquiera de las dos puertas; pero este no es el caso ya que el presentador puede descartar una puerta porque él sabe (a) qué puerta hemos seleccionado y (b) qué puerta tiene el coche.

```python
def elegir_puerta():
    """
    Función para elegir una puerta. Devuelve 1, 2, o 3 en forma aleatoria.
    """
    return np.random.randint(1,4)


class MontyHall:
    """
    Clase para modelar el problema de Monty Hall.
    """
    def __init__(self):
        """
        Crea la instancia del problema.
        """
        # Elige una puerta en forma aleatoria.
        self.puerta_ganadora = elegir_puerta()
        # variables para la puerta elegida y la puerta descartada
        self.puerta_elegida = None
        self.puerta_descartada = None

    def selecciona_puerta(self):
        """
        Selecciona la puerta del concursante en forma aleatoria.
        """
        self.puerta_elegida = elegir_puerta()

    def descarta_puerta(self):
        """
        Con este método el presentador descarta una de la puertas.
        """
        # elegir puerta en forma aleatoria .
        d = elegir_puerta()
        # Si es al puerta ganadora o la del concursante, volver a elegir.
        while d == self.puerta_ganadora or d == self.puerta_elegida:
            d = elegir_puerta()
        # Asignar el valor a puerta_descartada.
        self.puerta_descartada = d

    def cambiar_puerta(self):
        """
        Cambia la puerta del concursante una vez que se elimino una puerta.
        """
        # Solo existe una puerta para elegir.
        self.puerta_elegida = 6 - self.puerta_elegida - self.puerta_descartada

    def gana_concursante(self):
        """
        Determina si el concursante gana.
        Devuelve True si gana, False si pierde.
        """
        return self.puerta_elegida == self.puerta_ganadora


    def jugar(self, cambiar=True):
        """
        Una vez que la clase se inicio, jugar el concurso.

        'cambiar' determina si el concursante cambia su elección.
        """
        # El concursante elige una puerta.
        self.selecciona_puerta()
        # El presentador elimina una puerta.
        self.descarta_puerta()
        # El concursante cambia su elección.
        if cambiar:
            self.cambiar_puerta()
        # Determinar si el concursante ha ganado.
        return self.gana_concursante()

# Ahora, jugamos el concurso. primero nos vamos a quedar con nuestra elección inicial
# Vamos a ejecutar el experimiento 10.000 veces.gana, 
gana, pierde = 0, 0
N=1000
for i in range(N):
    # Crear la instancia del problema.
    s2 = MontyHall()
    # ejecutar el concurso sin cambiar de puerta..
    if s2.jugar(cambiar=False):
        # si devuelve True significa que gana.
        gana += 1
    else:
        # si devuelve False significa que pierde.
        pierde += 1

# veamos la fecuencia de victorias del concursante.
porc_gana = 100.0 * gana / (gana + pierde)

print("\n Concursos sin cambiar de puerta:",N)
print("  Gana: {0:} concursos".format(gana))
print("  Pierde: {0:} concursos".format(pierde))
print("  Porcentaje de victorias: {0:.6f} ".format(porc_gana))
```

Ahora, jugamos el concurso cambiando siempre la elección inicial

```python
# Vamos a ejecutar el experimiento 10.000 veces.
gana, pierde = 0, 0
N=10000
for i in range(N):
    # Crear la instancia del problema.
    s2 = MontyHall()
    # ejecutar el concurso sin cambiar de puerta..
    if s2.jugar(cambiar=True):
        # si devuelve True significa que gana.
        gana += 1
    else:
        # si devuelve False significa que pierde.
        pierde += 1

# veamos la fecuencia de victorias del concursante.
porc_gana = 100.0 * gana / (gana + pierde)

print("\n Concursos cambiando siempre de puerta:",N)
print("  Gana: {0:} concursos".format(gana))
print("  Pierde: {0:} concursos".format(pierde))
print("  Porcentaje de victorias: {0:.6f} ".format(porc_gana))
```

```python
def experimento_urnas():
    '''
    Función aleatoria que devuelve el resultado de simular el experimento *una vez*.
    El resultado es distinto cada vez que llamamos a la función.
    El resultado puede ser 'B' (bola blanca) o 'R' (bola roja)
    '''
    urna = random.randint(1,2)
    if urna == 1:
        #Escogemos una bola.
        return random.choice('BBBRR')
    else:
        #Escogemos una bola. Las bolas son [B,B,B,B,R,R]
        return random.choice('BBBBRR')

def es_bola_blanca(bola):
    '''
    Función que recibe como argumento un string, que puede ser
        "B" (representa una bola blanca)
        "R" (representa una bola roja)
    Devuelve True <=> el argumento es "B" (una bola blanca)
    '''
    if bola=='B':
        return True
    else:
        return False

random.seed(1) #Prueba a cambiar la semilla: la aproximación cambia 
N = 1000
muestra = [experimento_urnas() for _ in range(N)]
prob = sum(es_bola_blanca(e) for e in muestra)/N
print('Probabilidad aproximada de bola blanca:', prob)

```

El teorema de Bayes, es una proposición planteada por el matemático inglés Thomas Bayes y extendida por Pierre-Simon Laplace que expresa la probabilidad condicional de un evento aleatorio A dado B:

$$P(A|B) = \frac{P(A)P(B|A)}{P(B)}$$

Además contamos con la **versión de Laplace** donde contamos con $A_1, A_2,..., A_i,..., A_n$, una partición de espacio muestral, y sea $B$ un evento cualquiera (las mismas hipótesis del teorema de la probabilidad total). Se cumple:

$$P(A_i|B) = \frac{P(A_i)P(B|A_i)}{P(B)} = \frac{P(A_i)P(B|A_i)}{\sum_{j=1}^n P(B|A_j)P(A_j)}$$

Veamos como es utilizado el teorema con un ejemplo.

Suponga que en una visita a un consultorio del médico, una persona decide hacerse la prueba de detección de una enfermedad rara. Si tiene la mala suerte de recibir un resultado positivo, la siguiente pregunta lógica es: **"Dado el resultado de la prueba, ¿cuál es la probabilidad de que realmente tenga esta enfermedad?"**. El teorema de Bayes nos dice exactamente cómo calcular esta probabilidad:

$$P(E|+)=\frac{P(+|E)P(E)}{P(+)}$$

Como indica la ecuación, la probabilidad posterior de tener la enfermedad (E) dado que la prueba fue positiva (+) depende de la probabilidad a priori de la enfermedad $P(E)$. Imaginemos que

$$P(E) = 0.1$$
$$P(No E) = 1- P(E) = P(S) = 0.9$$

Además, la probabilidad posterior depende de la precisión de la prueba y para ello se necesita la frecuencia con la que la prueba informa correctamente un resultado negativo para un paciente sano, la frecuencia con la que la prueba informa de un resultado positivo para alguien con la enfermedad, etc. Es decir:

$$P(-|S) = 0.75$$
$$P(+|S) = 0.25$$
$$P(-|E) = 0.25$$
$$P(+|E) = 0.75$$

Finalmente, necesitamos saber la probabilidad general de un resultado positivo $P(+)$.


```python
import matplotlib.pyplot as plt

# Prior
P_E = 0.1
P_S = 0.9

# Likelihood
P_neg_S = 0.75
P_pos_S = 0.25
P_neg_E = 0.25
P_pos_E = 0.75

# Const
P_pos = P_pos_S*P_S + P_pos_E*P_E
P_neg = P_neg_S*P_S + P_neg_E*P_E

widths = [P_E, P_S]
indices = [widths[0]/2., widths[0] + widths[1]/2.]

plt.bar(indices, [1,1], widths, label="Negativo (-)")
plt.bar(indices, [P_pos_E,P_pos_S], widths, label="Positivo (+)")
plt.legend()
plt.yticks([0, P_pos_E, P_pos_S, 1], [0, P_pos_E, P_pos_S, 1])
plt.xticks(indices + [0, P_E,  1], ['P(E)', 'P(S)']+[0, P_E, 1])
plt.axvline(x=P_E, linewidth=1, color='r')
plt.show()
```

Finalmente podemos obtener la probabilidad posterior utilizando la fórmula

```python
P_E_pos = P_pos_E*P_E/P_pos
P_E_neg = P_neg_E*P_E/P_neg
P_S_pos = P_pos_S*P_S/P_pos
P_S_neg = P_neg_S*P_S/P_neg

print(P_E_pos, P_S_pos)

print(P_E_neg, P_S_neg)
```

Dado el resultado positivo de la prueba, la probabilidad de que realmente tenga esta enfermedad es solo un **25%**.

```python
def probabilidad_suma_impar_con_reposicion():
    """
    Calcula la probabilidad de que al sacar 2 bolas numeradas del 1 al 5 
    con reposición, la suma sea impar.
    """
    # Crear todas las posibles combinaciones
    combinaciones = []
    for i in range(1, 6):
        for j in range(1, 6):
            combinaciones.append((i, j))
    
    # Contar casos favorables (suma impar)
    casos_favorables = [(i, j) for i, j in combinaciones if (i + j) % 2 == 1]
    
    # Calcular probabilidad
    probabilidad = len(casos_favorables) / len(combinaciones)
    
    return probabilidad, casos_favorables, combinaciones

prob, favorables, totales = probabilidad_suma_impar_con_reposicion()
print(f"La probabilidad de que la suma sea impar (con reposición) es: {prob}")
print(f"En forma de fracción: {len(favorables)}/{len(totales)} = 13/25")
print(f"En porcentaje: {prob * 100}%")
print()
print("Casos favorables (algunos ejemplos):")
for caso in favorables[:5]:
    print(f"  {caso[0]} + {caso[1]} = {caso[0] + caso[1]}")
print("...")
```

```python
def probabilidad_suma_impar_sin_reposicion():
    """
    Calcula la probabilidad de que al sacar 2 bolas numeradas del 1 al 5 
    sin reposición, la suma sea impar.
    """
    # Crear todas las posibles combinaciones (sin repetir bolas)
    combinaciones = []
    for i in range(1, 6):
        for j in range(1, 6):
            if i != j:  # Sin reposición, no podemos tomar la misma bola dos veces
                combinaciones.append((i, j))
    
    # Contar casos favorables (suma impar)
    casos_favorables = [(i, j) for i, j in combinaciones if (i + j) % 2 == 1]
    
    # Calcular probabilidad
    probabilidad = len(casos_favorables) / len(combinaciones)
    
    return probabilidad, casos_favorables, combinaciones

prob, favorables, totales = probabilidad_suma_impar_sin_reposicion()
print(f"La probabilidad de que la suma sea impar (sin reposición) es: {prob}")
print(f"En forma de fracción: {len(favorables)}/{len(totales)} = 12/20 = 3/5")
print(f"En porcentaje: {prob * 100}%")
print()
print("Casos favorables (algunos ejemplos):")
for caso in favorables[:5]:
    print(f"  {caso[0]} + {caso[1]} = {caso[0] + caso[1]}")
print("...")
```

```python
import math

def combinatoria(n, k):
    """
    Calcula el coeficiente binomial (n k)
    """
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

def probabilidad_tres_caras_cinco_lanzamientos():
    """
    Calcula la probabilidad de sacar exactamente tres caras en cinco lanzamientos
    de moneda usando el coeficiente binomial.
    """
    # Número total de resultados posibles en 5 lanzamientos: 2^5 = 32
    total_casos = 2**5
    
    # Calcular el número de formas de obtener exactamente 3 caras en 5 lanzamientos
    formas_3_caras = combinatoria(5, 3)
    
    # Cada forma tiene probabilidad (1/2)^3 * (1/2)^2 = (1/2)^5
    probabilidad_forma = (1/2)**5
    
    # Probabilidad total: formas * probabilidad de cada forma
    probabilidad = formas_3_caras * probabilidad_forma
    
    return probabilidad, formas_3_caras, total_casos

prob, formas, total = probabilidad_tres_caras_cinco_lanzamientos()
print(f"La probabilidad de sacar exactamente 3 caras en 5 lanzamientos es: {prob}")
print(f"Hay {formas} formas de obtener 3 caras en 5 lanzamientos")
print(f"En forma de fracción: {formas}/{total} = 10/32 = 5/16")
print(f"En porcentaje: {prob * 100}%")
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def experimento_bolas_con_reposicion(n_simulaciones=1000):
    """
    Simula n veces el experimento de tomar 5 bolas con reposición de una caja con:
    - 10 bolas blancas
    - 20 bolas rojas
    - 30 bolas verdes
    
    Calcula la probabilidad de:
    a) Tomar 3 blancas y 2 rojas
    b) Tomar todas del mismo color
    """
    # Definir colores y probabilidades
    colores = ['blanca', 'roja', 'verde']
    probabilidades = [10/60, 20/60, 30/60]  # Probabilidad de cada color
    
    # Contadores
    contador_3b_2r = 0
    contador_mismo_color = 0
    
    # Resultados de cada simulación
    resultados = []
    
    for _ in range(n_simulaciones):
        # Realizar una simulación tomando 5 bolas
        bolas = np.random.choice(colores, size=5, p=probabilidades)
        resultados.append(bolas)
        
        # Contar bolas de cada color
        n_blancas = np.sum(bolas == 'blanca')
        n_rojas = np.sum(bolas == 'roja')
        n_verdes = np.sum(bolas == 'verde')
        
        # Verificar condición a) 3 blancas y 2 rojas
        if n_blancas == 3 and n_rojas == 2:
            contador_3b_2r += 1
        
        # Verificar condición b) todas del mismo color
        if n_blancas == 5 or n_rojas == 5 or n_verdes == 5:
            contador_mismo_color += 1
    
    # Calcular probabilidades
    prob_3b_2r = contador_3b_2r / n_simulaciones
    prob_mismo_color = contador_mismo_color / n_simulaciones
    
    return prob_3b_2r, prob_mismo_color, resultados

# Ejecutar simulación
prob_a, prob_b, resultados = experimento_bolas_con_reposicion(1000)

print(f"Basado en 1000 simulaciones:")
print(f"a) Probabilidad de tomar 3 blancas y 2 rojas: {prob_a:.4f}")
print(f"b) Probabilidad de tomar todas del mismo color: {prob_b:.4f}")

# Calcular valor teórico para comparar (usando combinatoria)
prob_teorica_a = combinatoria(5, 3) * ((10/60)**3) * ((20/60)**2)
prob_teorica_b = (10/60)**5 + (20/60)**5 + (30/60)**5

print("\nValores teóricos:")
print(f"a) Probabilidad teórica de tomar 3 blancas y 2 rojas: {prob_teorica_a:.4f}")
print(f"b) Probabilidad teórica de tomar todas del mismo color: {prob_teorica_b:.4f}")
```

```python
def prob_k_caras_n_lanzamientos(n, k):
    """
    Calcula la probabilidad de obtener exactamente k caras en n lanzamientos
    de una moneda justa.
    
    Usa la fórmula: (n k) * (1/2)^k * (1/2)^(n-k) = (n k) * (1/2)^n
    
    Parámetros:
    n -- número de lanzamientos
    k -- número de caras deseado
    
    Retorna:
    La probabilidad exacta
    """
    # Calcular coeficiente binomial (n k)
    coef = combinatoria(n, k)
    
    # Calcular probabilidad
    prob = coef * (1/2)**n
    
    return prob

# Calcular y mostrar las probabilidades para 0, 1, 2, 3, 4 y 5 caras en 5 lanzamientos
n_lanzamientos = 5
resultados = []

for k in range(n_lanzamientos + 1):
    prob = prob_k_caras_n_lanzamientos(n_lanzamientos, k)
    resultados.append((k, prob))
    print(f"Probabilidad de obtener exactamente {k} caras en {n_lanzamientos} lanzamientos: {prob:.6f}")

# Verificar que la suma de todas las probabilidades es 1
suma_probs = sum(prob for _, prob in resultados)
print(f"\nLa suma de todas las probabilidades es: {suma_probs}")

# Visualizar las probabilidades
k_valores = [k for k, _ in resultados]
probs = [prob for _, prob in resultados]

plt.figure(figsize=(10, 6))
plt.bar(k_valores, probs)
plt.title('Distribución de probabilidad: Número de caras en 5 lanzamientos')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.xticks(k_valores)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def probabilidad_suceso_y(prob_x, prob_al_menos_uno):
    """
    Calcula la probabilidad del suceso Y, dados:
    - prob_x: probabilidad de ocurrencia del suceso X
    - prob_al_menos_uno: probabilidad de ocurrencia de al menos un suceso
    
    Si X e Y son independientes, entonces:
    P(X ∪ Y) = P(X) + P(Y) - P(X)P(Y)
    
    Despejando P(Y):
    P(Y) = [P(X ∪ Y) - P(X)] / [1 - P(X)]
    """
    # P(X ∪ Y) es la probabilidad de que ocurra al menos un suceso
    p_x_union_y = prob_al_menos_uno
    
    # Para sucesos independientes:
    # P(X ∩ Y) = P(X)P(Y)
    # P(X ∪ Y) = P(X) + P(Y) - P(X)P(Y)
    
    # Despejar P(Y)
    p_y = (p_x_union_y - prob_x) / (1 - prob_x)
    
    return p_y

# Datos del problema
p_x = 0.4
p_al_menos_uno = 0.6

# Calcular P(Y)
p_y = probabilidad_suceso_y(p_x, p_al_menos_uno)

print(f"La probabilidad del suceso X es: {p_x}")
print(f"La probabilidad de que ocurra al menos un suceso es: {p_al_menos_uno}")
print(f"La probabilidad del suceso Y es: {p_y:.4f}")

# Verificar que la respuesta es correcta:
p_x_union_y_calculado = p_x + p_y - p_x * p_y
print(f"\nVerificación: P(X ∪ Y) = {p_x_union_y_calculado:.4f}")
```

```python
def prob_urna_a_dado_roja():
    """
    Calcula la probabilidad de que una bola roja haya sido extraída de la urna A
    usando el teorema de Bayes.
    
    Datos:
    - Urna A: 3 bolas rojas y 5 negras
    - Urna B: 2 bolas rojas y 1 negra
    - Urna C: 2 bolas rojas y 3 negras
    """
    # Probabilidades a priori de elegir cada urna
    p_urna_a = 1/3
    p_urna_b = 1/3
    p_urna_c = 1/3
    
    # Probabilidades de extraer una bola roja de cada urna
    p_roja_dado_a = 3/8
    p_roja_dado_b = 2/3
    p_roja_dado_c = 2/5
    
    # Probabilidad total de extraer una bola roja
    p_roja = p_urna_a * p_roja_dado_a + p_urna_b * p_roja_dado_b + p_urna_c * p_roja_dado_c
    
    # Probabilidad de que la bola haya sido extraída de la urna A, dado que es roja
    p_a_dado_roja = (p_urna_a * p_roja_dado_a) / p_roja
    
    return p_a_dado_roja, p_roja

prob, p_roja = prob_urna_a_dado_roja()
print(f"La probabilidad de que la bola roja haya sido extraída de la urna A es: {prob:.4f}")
print(f"La probabilidad total de extraer una bola roja es: {p_roja:.4f}")
```

```python
def prob_puerta_defectuosa():
    """
    Calcula la probabilidad de que una puerta elegida al azar sea defectuosa.
    
    Datos:
    - Grupo A: 380 puertas, 4% defectuosas
    - Grupo B: 270 puertas, 3% defectuosas
    - Grupo C: 350 puertas, 6% defectuosas
    """
    # Datos de los grupos
    grupo_a = {'total': 380, 'porcentaje_defectuosas': 0.04}
    grupo_b = {'total': 270, 'porcentaje_defectuosas': 0.03}
    grupo_c = {'total': 350, 'porcentaje_defectuosas': 0.06}
    
    # Calcular número de puertas defectuosas en cada grupo
    grupo_a['defectuosas'] = grupo_a['total'] * grupo_a['porcentaje_defectuosas']
    grupo_b['defectuosas'] = grupo_b['total'] * grupo_b['porcentaje_defectuosas']
    grupo_c['defectuosas'] = grupo_c['total'] * grupo_c['porcentaje_defectuosas']
    
    # Total de puertas
    total_puertas = grupo_a['total'] + grupo_b['total'] + grupo_c['total']
    
    # Total de puertas defectuosas
    total_defectuosas = grupo_a['defectuosas'] + grupo_b['defectuosas'] + grupo_c['defectuosas']
    
    # Probabilidad de que una puerta sea defectuosa
    prob_defectuosa = total_defectuosas / total_puertas
    
    return prob_defectuosa, grupo_a, grupo_b, grupo_c, total_puertas, total_defectuosas

prob, grupo_a, grupo_b, grupo_c, total, defectuosas = prob_puerta_defectuosa()

print(f"Grupo A: {grupo_a['total']} puertas, {grupo_a['defectuosas']} defectuosas ({grupo_a['porcentaje_defectuosas']*100}%)")
print(f"Grupo B: {grupo_b['total']} puertas, {grupo_b['defectuosas']} defectuosas ({grupo_b['porcentaje_defectuosas']*100}%)")
print(f"Grupo C: {grupo_c['total']} puertas, {grupo_c['defectuosas']} defectuosas ({grupo_c['porcentaje_defectuosas']*100}%)")
print(f"Total: {total} puertas, {defectuosas} defectuosas")
print(f"\nLa probabilidad de que una puerta elegida al azar sea defectuosa es: {prob:.4f}")
print(f"En porcentaje: {prob*100:.2f}%")
```
