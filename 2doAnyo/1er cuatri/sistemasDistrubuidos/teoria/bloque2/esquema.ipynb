## **1. Conceptos Básicos**

### **1.1 Dispositivos**

#### **Definición**
Los dispositivos de red son componentes físicos que facilitan la comunicación y el intercambio de datos dentro de una red informática.

#### **Categorías Principales**

- **Dispositivos Finales:**
  - **Definición:** Dispositivos que los usuarios utilizan para interactuar directamente con la red.
  - **Funciones:** Pueden funcionar como Cliente, Servidor o Cliente/Servidor.
  - **Ejemplos:**
    - Impresoras
    - Escáneres
    - Servidores de red
    - Equipos de correo de sobremesa
    - Tabletas
    - Teléfonos inteligentes
    - Cámaras IP
    - Teléfonos de vídeo
    - Dispositivos IoT (Internet de las Cosas)
    - Dispositivos de almacenamiento

- **Dispositivos Intermedios:**
  - **Definición:** Dispositivos que facilitan la comunicación y la gestión del tráfico de datos entre dispositivos finales.
  - **Ejemplos:**
    - Repetidores
    - Hubs
    - Puentes (Bridges)
    - Switches
    - Routers
    - Firewalls
    - Proxies

#### **1.1.1 Dispositivos Finales vs Intermedios**
- **Dispositivos Finales:**
  - **Ejemplos Específicos y Funciones:**
    - **Impresora, Escáner, etc.:** Utilizados para tareas específicas como imprimir o escanear documentos.
    - **Servidor de Red:** Gestiona recursos y servicios para otros dispositivos en la red.
    - **Equipos de Correo de Sobremesa:** Facilitan la comunicación por correo electrónico.
    - **Tabletas y Teléfonos Inteligentes:** Dispositivos móviles con capacidades de red.
    - **Cámaras IP y Teléfonos de Vídeo:** Dispositivos multimedia conectados a la red.
    - **Dispositivos IoT:** Objetos conectados a la red que recopilan y transmiten datos.
    - **Dispositivos de Almacenamiento:** Soluciones de almacenamiento en red como NAS.

- **Dispositivos Intermedios:**
  - **Ejemplos Específicos y Funciones:**
    - **Repetidor:** Amplifica y retransmite señales de red para extender su alcance.
    - **Hub:** Conecta múltiples dispositivos en una LAN, retransmitiendo paquetes a todos los puertos.
    - **Puente (Bridge):** Conecta y filtra tráfico entre segmentos de red a nivel de enlace de datos.
    - **Switch:** Conecta múltiples dispositivos en una LAN y gestiona el tráfico de datos enviando paquetes únicamente al dispositivo de destino utilizando direcciones MAC.
    - **Router:** Dirige el tráfico de datos entre diferentes redes, determinando las mejores rutas.
    - **Firewall:** Controla el tráfico de red mediante reglas para proteger redes internas de accesos no autorizados.
    - **Proxy:** Actúa como intermediario para solicitudes de recursos o servicios externos.

### **1.1.2 Dispositivos Específicos**

- **Repetidor:**
  - **Función:** Amplifica y retransmite señales de red para extender su alcance.
  - **Operación:** A nivel físico (Capa 1 del Modelo OSI), sin interpretación o modificación de los datos.

- **Hub:**
  - **Función:** Conecta varios dispositivos en una red local (LAN) y retransmite paquetes a todos los puertos.
  - **Operación:** Actúa como un punto de conexión central, operando a nivel físico (Capa 1).

- **Puente (Bridge):**
  - **Función:** Conecta y filtra tráfico entre segmentos de red a nivel de enlace de datos (Capa 2).
  - **Diferencia con Repetidor:** Filtra y redirige tráfico basado en direcciones MAC.

- **Switch:**
  - **Función:** Conecta múltiples dispositivos en una LAN y gestiona el tráfico de datos enviando paquetes únicamente al dispositivo de destino.
  - **Características:** Mantiene una tabla de direcciones MAC.
  - **Operación:** A nivel de enlace de datos (Capa 2).

- **Router/Gateway:**
  - **Función:** Dirige el tráfico de datos entre redes diferentes, adaptando protocolos y determinando la mejor ruta (Capa 3).
  - **Gateway:** Routers que actúan como punto de entrada y salida en una red.

- **Firewall:**
  - **Función:** Sistema de seguridad que controla el tráfico de red mediante reglas, protegiendo las redes internas de accesos no autorizados.

- **Proxy:**
  - **Función:** Actúa como intermediario para solicitudes de recursos o servicios externos.
  - **Tipos de Proxy:**
    - **Proxy Caché:** Almacena información de servicios web para mejorar el rendimiento y reducir el tráfico de red.
    - **Proxy Inverso:** Intermediario entre servicios internos y el exterior, generalmente actuando como balanceador de carga.

### **1.2 Medios**

#### **Definición**
Canales físicos a través de los cuales se transmiten los datos (mensajes) entre los dispositivos en una red.

#### **Tipos de Medios**
- **Cables de Cobre:**
  - **Coaxial**
  - **UTP/STP (Unshielded/Shielded Twisted Pair)**
- **Fibra Óptica:**
  - Transmite datos mediante ondas de luz o láser.
- **Inalámbrico:**
  - Utiliza ondas de radio o microondas para la transmisión de datos.

### **1.3 Interfaz de Red (NIC)**

#### **Definición**
Hardware adicional que permite a un dispositivo conectarse a una red física a través de un medio concreto.

#### **Tipos de Interfaces**
- **Para Cables de Cobre:**
  - Bus ISA
  - Bus PCI
  - Bus PCI Express
- **Para Fibra Óptica:**
  - Interfaces específicas según el tipo de fibra.
- **Inalámbricas:**
  - USB, etc.

### **1.4 Topologías**

#### **Definición**
Las topologías de red describen la disposición física o lógica de los dispositivos en una red de computadores. Afectan al rendimiento, la eficiencia y la escalabilidad de la red.

#### **Principales Topologías**
- **Bus**
- **Estrella**
- **Anillo**
- **Doble Anillo**
- **Árbol**
- **Malla**
- **Malla Conexa**
- **Mixta**

#### **Topología Física vs Lógica**
- **Física:** Cómo se conectan físicamente los dispositivos.
- **Lógica:** Cómo se comunican los dispositivos.

#### **Ejemplos de Topologías Ethernet**
- **Red Ethernet 10Base2:**
  - **Topología Física:** Bus
  - **Topología Lógica:** Bus
- **Red Ethernet 10Base-T con Hub:**
  - **Topología Física:** Estrella
  - **Topología Lógica:** Bus

### **1.5 Tipos de Redes**

#### **Red de Área Local (LAN)**
- **Definición:** Red utilizada en un hogar, una empresa, un edificio o un campus en un área geográfica reducida.
- **Características:**
  - Alta velocidad de transmisión.
  - Controlada y gestionada por una sola entidad.
  - Uso de dispositivos como switches y routers.

#### **Red de Área Extensa (WAN)**
- **Definición:** Cubre áreas geográficas extensas como ciudades, países o incluso continentes.
- **Características:**
  - Conecta diferentes LAN.
  - Utiliza tecnologías como MPLS, Frame Relay, etc.
  - Generalmente gestionada por múltiples entidades.

### **1.6 Reglas y Protocolos**

#### **Definición de Protocolo**
Conjunto de reglas predeterminadas para la comunicación. Define cómo identificar dispositivos, cómo transmitir datos y cómo procesar información.

#### **Pila de Protocolos**
Organiza los protocolos en capas, donde cada capa representa una abstracción y proporciona servicios a las capas adyacentes.

#### **Modelo de Referencia: Modelo OSI**
1. **Capa Física**
2. **Capa de Enlace de Datos**
3. **Capa de Red**
4. **Capa de Transporte**
5. **Capa de Sesión**
6. **Capa de Presentación**
7. **Capa de Aplicación**

#### **Equivalencia entre Modelo OSI y Pila TCP/IP**

| **Capas del Modelo OSI** | **Capas de la Pila TCP/IP** |
|--------------------------|------------------------------|
| 7. Aplicación            | Aplicación                   |
| 6. Presentación          | Aplicación                   |
| 5. Sesión                | Aplicación                   |
| 4. Transporte            | Transporte                   |
| 3. Red                   | Internet                     |
| 2. Enlace de Datos       | Acceso a la Red              |
| 1. Física                | Física                       |

#### **Protocolos TCP/IP**
- **Capa de Aplicación:**
  - **HTTP (HyperText Transfer Protocol):** Navegación web.
  - **FTP (File Transfer Protocol):** Transferencia de archivos.
  - **DNS (Domain Name System):** Resolución de nombres de dominio.
  - **SMTP (Simple Mail Transfer Protocol):** Envío de correos electrónicos.
  - **SSH (Secure Shell):** Acceso remoto seguro.
  - **NFS (Network File System):** Sistemas de archivos en red.

- **Capa de Transporte:**
  - **TCP (Transmission Control Protocol):** Comunicación orientada a conexión con garantías.
  - **UDP (User Datagram Protocol):** Comunicación no orientada a conexión sin garantías.

- **Capa de Internet:**
  - **IP (Internet Protocol):** Direccionamiento y encaminamiento de paquetes.
  - **ICMP (Internet Control Message Protocol):** Mensajes de control y error (e.g., ping).

- **Capa de Acceso a la Red:**
  - **Ethernet, Fast Ethernet, WiFi, SLIP & PPP, FDDI, Frame Relay, Proxy:** Protocolos y tecnologías para la transmisión de datos en la red física.

## **2. Red Física**

### **2.1 Definiciones Básicas**

#### **Red Física**
Infraestructura tangible que facilita la conexión y comunicación entre dispositivos en una red de computadores. Se especifica en las primeras capas del modelo OSI y define cómo serán las conexiones físicas entre los dispositivos de la red (nodos).

#### **Componentes de la Red Física**
- **Medios:** Cables de cobre, fibra óptica, inalámbricos.
- **Topología:** Disposición de los dispositivos en la red.
- **Electrónica:** Equipos necesarios para la transmisión de señales.

#### **Terminología**
- **Trama (Frame):** Unidad de datos de la capa de enlace de datos.
- **Bit:** Unidad de datos para la capa física.
- **Nodo:** Dispositivo de red conectado a un medio común.
- **Red Física:** Conjunto de dos o más nodos conectados a un medio común.

#### **Ancho de Banda**
- **Definición:** Capacidad que posee un medio para transportar datos sin procesar.
- **Medición:** Kilobits por segundo (kbps), Megabits por segundo (Mbps), Gigabits por segundo (Gbps).
- **Factores que Afectan el Ancho de Banda:**
  - Propiedades de las tecnologías y los medios físicos utilizados para la señalización y detección de señales de red.

#### **Rendimiento Real**
- **Definición:** Medida de transferencia de bits (velocidad) a través de los medios durante un período de tiempo determinado.
- **Características:**
  - Generalmente inferior al ancho de banda especificado.
  - Afectado por retrasos introducidos por dispositivos intermedios.
  - Calculado aplicando teoría de colas.

#### **Capacidad de Transferencia Útil**
- **Definición:** Medida de datos utilizables transferidos durante un período de tiempo determinado en bps.
- **Importancia:** Representa la velocidad con la que se puede transferir un archivo entre dos equipos.
- **Medición:** bps como cociente entre el tamaño del archivo y el tiempo de transferencia.

#### **Resumen Capacidad de Transmisión**
- **Ancho de Banda:** Velocidad de transferencia máxima teórica.
- **Rendimiento:** Velocidad de transferencia real.
- **Capacidad de Transferencia Útil:** Velocidad de transferencia de los datos.

### **2.2 Capas OSI Relacionadas**

#### **Capa Física (Capa 1)**
- **Función:** Maneja la transmisión de bits a través de un medio físico.
- **Características:**
  - Define voltaje, velocidad de transmisión, y aspectos físicos de la señalización.
  - Ejemplos de dispositivos: Repetidores, Hubs.

#### **Capa de Enlace de Datos (Capa 2)**
- **Subcapas:**
  - **Media Access Control (MAC):** Controla cómo los dispositivos acceden al medio físico y gestionan las direcciones MAC.
  - **Logical Link Control (LLC):** Gestiona la comunicación entre dispositivos en la misma red.
- **Función:** Define cómo se inserta y recupera la información del medio (tramas).
- **Ejemplos de dispositivos:** Puentes, Switches.

#### **Dispositivos y Capas OSI**
- **Capa Física:**
  - Repetidores, Hubs.
- **Capa de Enlace de Datos:**
  - Puentes, Switches.
- **Capa de Red:**
  - Routers.

### **2.3 VLAN (Red de Área Local Virtual)**

#### **Definición**
VLAN (Virtual Local Area Network) es un agrupamiento lógico de dispositivos de una LAN por su función, departamento, aplicación, etc., independientemente de la ubicación física de su segmento de LAN.

#### **Funcionamiento**
- **Etiquetado de Tramas:** Cada trama se etiqueta con el ID de la VLAN utilizando la etiqueta IEEE 802.1Q.
- **Configuración:** Se configuran mediante software en switches gestionados.

#### **Ventajas**
- **Mejora del Desempeño:** Agrupando usuarios y recursos que se comunican con mayor frecuencia.
- **Escalabilidad:** Facilita la expansión de la red sin cambios físicos.
- **Seguridad:** Aislamiento de segmentos de red para diferentes departamentos o funciones.
- **Facilidad de Gestión:** Simplifica la administración de la red.

#### **Implementación**
- **Switch Gestionado:** Permite la configuración de múltiples VLANs y la gestión de etiquetado.
- **Segmentación de la LAN:** Comparación entre una LAN tradicional y una LAN con VLAN gestionadas.

#### **Ejemplo de Configuración**
- **LAN Tradicional:**
  ```
  [Switch] --- [Router] --- [Internet]
  ```
- **LAN con VLAN Gestionadas:**
  ```
  [Switch Gestionado] --- [Router]
     |         |         |
  VLAN 1    VLAN 2    VLAN 3
  ```

## **3. TCP/IP: Internet**

### **3.1 Introducción**

#### **Definición**
TCP/IP (Transmission Control Protocol/Internet Protocol) es un conjunto de protocolos de red en los que se basa Internet.

#### **Componentes Principales**
- **TCP (Transmission Control Protocol):** Protocolo de Transmisión.
- **IP (Internet Protocol):** Protocolo de Internet.

#### **Historia**
- **Desarrollo:** TCP/IP fue desarrollado en 1973 por Vinton Cerf como parte de un proyecto de ARPA (Agencia de Proyectos de Investigación Avanzada).
- **ARPANet:** Inicialmente, Internet comenzó como ARPANet, una red informática que conectaba redes de universidades y laboratorios de investigación en USA desde 1966.
- **Adopción Oficial:** En 1983, TCP/IP fue adoptado oficialmente por ARPANet.
- **World Wide Web (WWW):** Desarrollada en 1989 por el CERN (Consejo Europeo de Investigación Nuclear).

### **3.2 Pila de Protocolos**

#### **Comparación TCP/IP vs Modelo OSI**

| 6. Presentación          | Aplicación                   |
| 5. Sesión                | Aplicación                   |
| 4. Transporte            | Transporte                   |
| 3. Red                   | Internet                     |
| 2. Enlace de Datos       | Acceso a la Red              |
| 1. Física                | Física                       |

#### **Protocolos Principales en TCP/IP**
- **Capa de Aplicación:**
  - **HTTP (HyperText Transfer Protocol):** Navegación web.
  - **FTP (File Transfer Protocol):** Transferencia de archivos.
  - **DNS (Domain Name System):** Resolución de nombres de dominio.
  - **SMTP (Simple Mail Transfer Protocol):** Envío de correos electrónicos.
  - **SSH (Secure Shell):** Acceso remoto seguro.
  - **NFS (Network File System):** Sistemas de archivos en red.

- **Capa de Transporte:**
  - **TCP (Transmission Control Protocol):** Comunicación orientada a conexión con garantías.
  - **UDP (User Datagram Protocol):** Comunicación no orientada a conexión sin garantías.

- **Capa de Internet:**
  - **IP (Internet Protocol):** Direccionamiento y encaminamiento de paquetes.
  - **ICMP (Internet Control Message Protocol):** Mensajes de control y error (e.g., ping).

- **Capa de Acceso a la Red:**
  - **Ethernet, Fast Ethernet, WiFi, SLIP & PPP, FDDI, Frame Relay, Proxy:** Protocolos y tecnologías para la transmisión de datos en la red física.

### **3.3 Unidad de Datos (PDU)**

#### **Proceso de Encapsulación**
- **Aplicación (Datos) → Transporte (Segmento) → Internet (Paquete/Datagrama) → Acceso a la Red (Trama) → Red Física (Señales).**

#### **Detalles del Proceso**
1. **Aplicación:**
   - Los datos generados por la aplicación son la unidad de datos inicial.
2. **Transporte:**
   - Los datos de la aplicación se segmentan y se añaden cabeceras TCP/UDP.
3. **Internet:**
   - Los segmentos se encapsulan en paquetes o datagramas IP.
4. **Acceso a la Red:**
   - Los paquetes IP se encapsulan en tramas para la transmisión física.
5. **Red Física:**
   - Las tramas se convierten en señales eléctricas, ópticas o de radio para su transmisión.

### **3.4 Direcciones**

#### **Tipos de Direcciones**
- **Dirección MAC (Media Access Control):**
  - **Definición:** Identificador único asignado a cada interfaz de red en hardware (Capa 2).
  - **Formato:** 48 bits, expresados en hexadecimal (e.g., 00:1A:2B:3C:4D:5E).
  - **Características:** Física y fija, quemada en el hardware durante la fabricación.

- **Dirección IP:**
  - **Definición:** Etiqueta numérica que identifica de forma lógica un dispositivo en una red (Capa 3).
  - **Formato IPv4:** 32 bits, en 4 octetos separados por puntos (e.g., 192.168.0.1).
  - **Características:**
    - Incluye ID de red y ID de host mediante la máscara de subred.
    - Cada NIC debe tener una dirección IP única.
    - Las direcciones IP se asignan por una autoridad central (IANA).

- **Nombre de Dominio (DNS):**
  - **Definición:** Nombre jerárquico legible que se asocia a una dirección IP.
  - **Ejemplo:** www.dtic.ua.es ↔ 173.145.200.2.

- **Puerto:**
  - **Definición:** Número que identifica una aplicación específica dentro de un dispositivo en red.
  - **Formato:** Número de 16 bits (0-65535).
  - **Tipos:**
    - **Puertos Bien Conocidos (0-1023):** Reservados para servicios estándar (e.g., HTTP: 80, HTTPS: 443).
    - **Puertos Registrados (1024-49151):** Asignados para aplicaciones y servicios registrados (e.g., MySQL: 3306).
    - **Puertos Dinámicos o Privados (49152-65535):** Disponibles para uso general por aplicaciones temporales o personalizadas.

- **URL (Localizador Uniforme de Recursos):**
  - **Definición:** Especifica la dirección web o la ubicación de un recurso único en Internet.
  - **Ejemplos:**
    - https://www.ua.es:8080/sd/index.html?pag=2&col=5
    - https://192.168.0.4/index.html
    - ftp://ilorenzo@ftp.ua.es/sd/miDocumento.docx
    - ssh://ilorenzo:xh45jjxcd001b@ssh.ua.es/

#### **Asignación de Direcciones IP**

- **IANA (Internet Assigned Numbers Authority):**
  - Supervisa la asignación de espacios de direcciones IP a nivel global.
  - Delegación a organizaciones regionales (RIR).

- **RIR (Regional Internet Registries):**
  - **Ejemplos:**
    - **RIPE NCC:** Europa, Oriente Medio y partes de Asia Central.
    - **ARIN, APNIC, LACNIC, AFRINIC:** Otras regiones.
  - Distribuyen direcciones IP a nivel regional.

- **LIR (Local Internet Registries) e ISPs:**
  - Reciben bloques de direcciones IP de las RIR y los asignan a sus clientes.
  - **Proceso de Asignación:**
    1. **IANA a RIR**
    2. **RIR a LIR/ISP**
    3. **LIR/ISP a Usuarios Finales**

#### **Dirección IP y Máscara de Subred**

- **Estructura de la Dirección IPv4:**
  - **Formato:** 32 bits en 4 octetos (e.g., 192.168.0.1).
  - **Componentes:**
    - **ID de Red:** Determinado por la máscara de subred.
    - **ID de Host:** Parte restante para identificar dispositivos dentro de la red.

- **Máscara de Subred:**
  - **Definición:** Determina qué parte de la dirección IP identifica la red y qué parte al host.
  - **Formato:** Igual que una dirección IP (32 bits).
  - **Características:**
    - Los bits a 1 identifican la red.
    - Los bits a 0 identifican el host.
    - Deben estar agrupados al inicio (sin intercalados).
  - **Representación Comprimida:** Indica el número de bits a 1 (e.g., /24).

- **Ejemplo de Aplicación de la Máscara:**
  - **Dirección IP:** 128.223.254.10
  - **Máscara de Subred:** 255.255.255.0 (/24)
  - **Operaciones Lógicas:**
    - **Dirección de Red:** 128.223.254.0 (128.223.254.10 AND 255.255.255.0)
    - **Dirección de Host:** 0.0.0.10 (128.223.254.10 AND NOT 255.255.255.0)

#### **Rangos de Direcciones IP**

- **Direcciones Privadas:**
  - **Clase A:** 10.0.0.0 a 10.255.255.255 (/8)
  - **Clase B:** 172.16.0.0 a 172.31.255.255 (/16)
  - **Clase C:** 192.168.0.0 a 192.168.255.255 (/24)

- **Direcciones Reservadas:**
  - **Red 0.0.0.0:** Expresa conceptos como la ruta por defecto.
  - **Red 127.0.0.0:** Reservada para el loopback.
  - **Broadcast:** Última dirección de cada rango (e.g., 192.168.0.255).

#### **Subclasificación de Direcciones IP**

- **Ejemplo de Rango Clase C:**
  - **Rango:** 192.168.0.0/24
  - **Dirección de Red:** 192.168.0.0
  - **Rango de Hosts:** 192.168.0.1 a 192.168.0.254
  - **Broadcast:** 192.168.0.255
  - **Total de Hosts Útiles:** 254

- **Ejemplo con Submask /26:**
  - **Rango:** 192.168.0.0/26
  - **Dirección de Red:** 192.168.0.0
  - **Rango de Hosts:** 192.168.0.1 a 192.168.0.62
  - **Broadcast:** 192.168.0.63
  - **Total de Hosts Útiles:** 62

- **Ejemplo con Submask /25:**
  - **Rango:** 192.168.0.0/25
  - **Dirección de Red:** 192.168.0.0
  - **Rango de Hosts:** 192.168.0.1 a 192.168.0.126
  - **Broadcast:** 192.168.0.127
  - **Total de Hosts Útiles:** 126

### **3.5 Encaminamiento**

#### **Routers (Encaminadores)**
- **Definición:** Dispositivo de red intermedio conectado a dos o más redes.
- **Características:**
  - Requiere una NIC para cada conexión (cada una con su dirección IP).
  - Decide el trayecto de los paquetes que pasan por él.
- **Tipos de Envío:**
  - **Envío Directo:** El host envía a otro host que está en su propia red física.
  - **Envío Indirecto:** El destino del paquete IP está fuera de la red física, requiriendo la presencia de un router (Gateway).
- **Tipos de Paquetes IP:**
  - **Directo/Indirecto:** Diferenciación según si el destino está en la misma red o no.

#### **Ejemplo de Encaminamiento IP Directo e Indirecto**
- **Escenario:**
  - **Host A** en red X desea comunicarse con **Host Y** en red Y.
- **Proceso:**
  1. **Host A** envía un paquete dirigido a **Host Y**.
  2. **Switch** de la red X envía el paquete al **Router**.
  3. **Router** determina que **Host Y** está en una red diferente y envía el paquete hacia la red Y.
  4. **Switch** de la red Y recibe el paquete y lo envía a **Host Y**.

### **3.6 Transporte Seguro → SSL/TLS**

#### **SSL (Secure Sockets Layer) y TLS (Transport Layer Security)**
- **Definición:** Protocolo que agrega una capa de seguridad a la pila TCP/IP.
- **Funciones:**
  - **Cifrado Asimétrico:** Utilizado para la negociación de la conexión.
  - **Cifrado Simétrico:** Utilizado para la comunicación segura una vez establecida la conexión.
  - **Certificados Digitales:** Verificación de identidad y autenticidad.
- **Implementación:**
  - **HTTPS:** HTTP sobre SSL/TLS para comunicación web segura.

#### **Características de SSL/TLS**
- **Cifrado:** Protege la confidencialidad e integridad de los datos transmitidos.
- **Autenticación:** Verifica la identidad de los servidores y, opcionalmente, de los clientes.
- **Integridad:** Garantiza que los datos no han sido alterados durante la transmisión.

### **3.7 Sockets**

#### **Definición**
Biblioteca que proporciona la interfaz del Sistema Operativo para la comunicación de aplicaciones a través de redes utilizando TCP/IP.

#### **Componentes de un Socket**
- **Protocolo de Transporte:** TCP o UDP.
- **Dirección IP:** Dirección del host.
- **Puerto:** Número que identifica la aplicación específica.

#### **Funcionamiento de los Sockets**
- **Cliente/Servidor:**
  - **Cliente:** Crea un socket, se conecta al servidor, envía y recibe datos.
  - **Servidor:** Crea un socket, lo asocia a un puerto, escucha conexiones entrantes, acepta conexiones, y gestiona la comunicación con el cliente.

#### **Ejemplo de Cliente/Servidor HTTP con Sockets**
- **ClienteHTTP:**
  1. **Crear Socket:** `socket(AF_INET, SOCK_STREAM, 0)`
  2. **Conectar al Servidor:** `connect(sd, dirección IP, puerto)`
  3. **Enviar Solicitud HTTP:** `write(sd, "GET /recurso HTTP/1.0\r\n\r\n")`
  4. **Leer Respuesta HTTP:** `read(sd, buffer, tamaño)`
  5. **Cerrar Socket:** `close(sd)`

- **ServidorHTTP:**
  1. **Crear Socket:** `socket(AF_INET, SOCK_STREAM, 0)`
  2. **Asociar al Puerto:** `bind(sd, puerto)`
  3. **Escuchar Conexiones:** `listen(sd, backlog)`
  4. **Aceptar Conexiones Entrantes:** `accept(sd)`
  5. **Leer Solicitud HTTP:** `read(sdHijo, buffer, tamaño)`
  6. **Enviar Respuesta HTTP:** `write(sdHijo, "HTTP/1.0 200 OK\r\n\r\ncontenido")`
  7. **Cerrar Socket Hijo:** `close(sdHijo)`

### **3.8 IPv6**

#### **Situación Actual con IPv4**
- **Problemas de IPv4:**
  - Espacio de direcciones insuficiente (aproximadamente 4.000 millones).
  - Falta de soporte para nuevas aplicaciones.
  - Necesidad de identificar emisor y receptor.
  - Falta de QoS estándar.
  - Falta de seguridad integrada.

#### **Solución: IPv6**

#### **Características Principales de IPv6**
- **Compatibilidad con IPv4:** Implementación en pila dual.
- **Direcciones de 128 bits:** Proporciona un espacio de direcciones mucho mayor (≈3.4×10^38 direcciones).
- **Rediseño de los Datagramas:** Estructura más eficiente y flexible.
- **Seguridad Integrada:** Incorporación de IPSec para cifrado y autenticación.
- **Mejor Soporte para QoS:** Facilita la gestión de tráfico en tiempo real.
- **Preparado para Dispositivos Móviles y Redes P2P:** Mejora la movilidad y el enrutamiento.
- **Multicast:** Envío a múltiples destinatarios en una sola operación.
- **Implantación Parcial y Conjunta con IPv4:** Transición gradual desde IPv4.

### **3.9 Red Privada Virtual: VPN**

#### **Definición**
Una Red Privada Virtual (VPN) es una tecnología que permite establecer una conexión segura entre dos o más redes locales distribuidas geográficamente, creando túneles sobre Internet.

#### **Objetivo**
Garantizar la privacidad y la seguridad de los datos, permitiendo que usuarios remotos accedan a recursos de una red privada como si estuvieran físicamente en ella.

#### **Tipos de VPN Según Función y Escenario de Uso**
- **VPN de Acceso Remoto:**
  - **Descripción:** Conectan personas con organizaciones o redes domésticas.
  - **Usos Comunes:** Teletrabajo, acceso seguro desde ubicaciones remotas.

- **VPN de Sitio a Sitio:**
  - **Descripción:** Conectan de manera segura redes LAN separadas geográficamente.
  - **Usos Comunes:** Conexión entre sedes de una misma organización u organizaciones asociadas.
