```python
import pandas as pd
import matplotlib.pyplot as plt

TablaA1DF = pd.DataFrame({'fuerza': [1373.4,1275.3 ,1177.2 ,981,784.8 , 588.6 ,392.4 ,196.2 ],
                          'errorFuerza': [0.34,0.33,0.32,0.30,0.28,0.26,0.24,0.22],
                          'elongacion': [449,417,385,321,252,193,131,66],
                          'errorElongacion': [2,2,2,2,2,2,2,2]})
TablaA2DF = pd.DataFrame({'fuerza': [2452.5, 2256.3 ,1962,1765.8, 1471.5, 981, 784.8, 490.5 ],
                          'errorFuerza': [0.45, 0.43, 0.40, 0.38, 0.35, 0.30, 0.28, 0.25],
                            'elongacion': [201, 199, 176, 151, 131, 80, 51, 32],
                            'errorElongacion': [2, 2, 2, 2, 2, 2, 2, 2]})
def imprimir_grafica(DF,titulo, pendiente, intercepto):
    plt.figure(figsize=(10,7))
    plt.errorbar(DF['fuerza'],DF['elongacion'],yerr=DF['errorElongacion'],xerr=DF['errorFuerza'],fmt='o')
    plt.grid()
    plt.plot(DF['fuerza'],DF['elongacion'], color='red', linestyle='-', linewidth=1, label='Datos')
    plt.plot(DF['fuerza'],pendiente*DF['fuerza']+intercepto, color='blue', linestyle='-', linewidth=1, label=f'Ajuste de regresión ξ = {pendiente}·F + {intercepto}')
    plt.legend()
    plt.xlabel('Elongacion (mm)')
    plt.ylabel('Fuerza (N)')
    plt.title(titulo)
    plt.show()
# ξ = 0.33·F + 1.57
# ξ = 0.09·F - 12.6
imprimir_grafica(TablaA1DF,'Fuerza vs Elongacion A1',0.33,1.57)
imprimir_grafica(TablaA2DF,'Fuerza vs Elongacion A2',0.09,-12.6)
```
